```r
library(tidyverse)
library(rpart)
library(randomForest)
```

```r
###### Leitura da base ###### 
base_treino = readRDS(file="atividade/base_bike_treino_final.rds")
head(base_treino,2)
glimpse(base_treino)
```

```r
#fixar a semente
set.seed(123456789)
```

```r
#Ajuste do modelo
rf_cnt <- randomForest(x =  base_treino |>
                        select(-c(instant,
                                 dteday,
                                 casual,
                                 registered,
                                 cnt)), 
                      y = base_treino$cnt, ntree=500)
```

```r
#verificar se a função rodou direito
rf_cnt
```

```r
# Variáveis Importantes

importancia = rf_cnt$importance
sort(importancia[,1],decreasing = TRUE)
```

```r
# Previsão
base_teste = readRDS(file="atividade/base_bike_teste.rds")
predict(rf_cnt, newdata = base_teste)
```

```r
base_treino_class = base_treino |> 
  mutate(cnt_maior = ifelse(base_treino$cnt>=4000,"muitas bikes","poucas bikes")) |>
  select(-c(instant,
            dteday,
            casual,
            registered,
            cnt))
base_treino_class$cnt_maior = as.factor(base_treino_class$cnt_maior)
glimpse(base_treino_class)
```

```r
# Ajuste 

rf_cnt_maior <- randomForest(x = base_treino_class |> select(-cnt_maior),
                       y = base_treino_class$cnt_maior , 
                       ntree=500)

rf_cnt_maior

saveRDS(rf_cnt_maior,file = "atividade/rf_cnt_maior.rds")
```

```r
# Variáveis Importantes

importancia = rf_cnt_maior$importance
sort(importancia[,1],decreasing = TRUE)
```

```r
predict(rf_cnt_maior, newdata = base_teste)
```

```r
#### Multiclasses

base_treino_mult = base_treino |>
  mutate(cnt_mult_class = 
           ifelse(
             base_treino$cnt<4000,"poucas",
             ifelse(base_treino$cnt>4000 & base_treino$cnt<6000,"normal",
                    "muitas"))) |>
  select(-c(instant,
            dteday,
            casual,
            registered,
            cnt))
base_treino_mult$cnt_mult_class = as.factor(base_treino_mult$cnt_mult_class)

summary(base_treino_mult$cnt_mult_class)
```
